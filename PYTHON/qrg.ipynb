```python
import qrcode
from PIL import Image

def generate_qr_code(data, filename="qrcode1.png", color="#000000", background="#FFFFFF", embed_image_path=None):
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_H,
        box_size=10,
        border=4,
    )
    qr.add_data(data)
    qr.make(fit=True)

    # Create QR code image using PIL
    qr_img = qr.make_image(fill_color=color, back_color=background)

    if embed_image_path:
        # Open and resize the image
        embed_image = Image.open(embed_image_path)
        qr_size = qr_img.size[0]
        embed_image.thumbnail((qr_size/10, qr_size/10))

        # Calculate the position to paste the image (centered)
        position = ((qr_size - embed_image.size[0]) // 2, (qr_size - embed_image.size[1]) // 2)

        # Paste the image onto the QR code
        qr_img.paste(embed_image, position)

    # Save the image
    qr_img.save(filename)
    print(f"QR code has been generated and saved as '{filename}'.")

if __name__ == "__main__":
    # Example usage
    data = input("Enter the text or URL for the QR code: ")
    color = input("Enter the hex color for the QR code (default is #000000): ") or "#000000"
    background = input("Enter the hex background color for the QR code (default is #FFFFFF): ") or "#FFFFFF"
    embed_image_path = input("Enter the path of the image to embed in the QR code (or leave blank): ")

    generate_qr_code(data, color=color, background=background, embed_image_path=embed_image_path or None)
```

```python
import qrcode
from PIL import Image, ImageStat

def average_image_brightness(image):
    stat = ImageStat.Stat(image)
    r, g, b = stat.mean[:3]
    return (r + g + b) / 3

def adjust_qr_color(qr_img, base_img, dark_color=(0, 0, 0), light_color=(255, 255, 255)):
    qr_pixels = qr_img.load()
    base_pixels = base_img.load()
    width, height = qr_img.size

    for y in range(height):
        for x in range(width):
            base_color = base_pixels[x, y]
            qr_color = qr_pixels[x, y]
            
            # Calculate the brightness of the base image pixel
            brightness = sum(base_color[:3]) / 3

            if qr_color == (0, 0, 0, 255):  # black QR pixel
                # Interpolate between dark_color and light_color based on brightness
                interp_color = tuple(int(light_color[i] + (dark_color[i] - light_color[i]) * (1 - brightness / 255.0)) for i in range(3))
                qr_pixels[x, y] = interp_color + (255,)  # set alpha to 255
            else:
                qr_pixels[x, y] = (255, 255, 255, 0)  # make it transparent

    return qr_img

def create_custom_qr(data, color="#000000", background="#FFFFFF", bg_img_path=None, embed_image_path=None, qr_path='custom_qr.png'):
    if bg_img_path:
        # Load the base image
        base_img = Image.open(bg_img_path).convert("RGBA")
        img_w, img_h = base_img.size
        
        # Calculate the average brightness of the image
        brightness = average_image_brightness(base_img)

        # Set QR code color based on brightness if not provided
        if color == "#000000" :
            if brightness > 128:
                qr_color = "black"
            else:
                qr_color = "white"
        else:
            qr_color = color
        
        # Generate the QR code
        qr = qrcode.QRCode(
            version=1,
            error_correction=qrcode.constants.ERROR_CORRECT_H,
        )
        qr.add_data(data)
        qr.make(fit=True)
        
        # Create the QR code image
        qr_img = qr.make_image(fill_color=qr_color, back_color="transparent").convert("RGBA")
        qr_img = qr_img.resize((img_w, img_h))

        # Adjust QR code colors based on the base image
        qr_img = adjust_qr_color(qr_img, base_img, dark_color=(0, 0, 0), light_color=(255, 255, 255))

        # Composite the QR code onto the image
        combined_img = Image.alpha_composite(base_img, qr_img)
    else:
        # Generate the QR code with specified colors
        qr = qrcode.QRCode(
            version=1,
            error_correction=qrcode.constants.ERROR_CORRECT_H,
        )
        qr.add_data(data)
        qr.make(fit=True)
        
        # Create the QR code image using specified colors
        qr_img = qr.make_image(fill_color=color, back_color=background).convert("RGBA")
        combined_img = qr_img

    # Embed image in the center of the QR code if provided
    if embed_image_path:
        embed_image = Image.open(embed_image_path).convert("RGBA")
        qr_size = qr_img.size[0]
        embed_image.thumbnail((qr_size // 3, qr_size // 3))
        position = ((combined_img.size[0] - embed_image.size[0]) // 2, (combined_img.size[1] - embed_image.size[1]) // 2)
        combined_img.paste(embed_image, position, embed_image)

    # Save the final image
    combined_img.save(qr_path)
    print(f"Custom QR code saved as {qr_path}")

if __name__ == "__main__":
    data = input("Enter the data for the QR code: ")
    color = input("Enter the hex color for the QR code (default is #000000): ") or "#000000"
    background = input("Enter the hex background color for the QR code (default is #FFFFFF): ") or "#FFFFFF"
    bg_img_path = input("Enter the path of the background image (or leave blank): ")
    embed_image_path = input("Enter the path of the image to embed in the QR code (or leave blank): ")
    qr_path = input("Enter the output file path (default is 'custom_qr.png'): ") or 'custom_qr.png'

    create_custom_qr(data, color, background, bg_img_path or None, embed_image_path or None, qr_path)
```
